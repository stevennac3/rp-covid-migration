## Produce daily Origin-Destination trip matrix for NYC Destination borough using csv generated by EDM 

```python
import pandas as pd
import numpy as np
import os
import time
```

```python
from geo import stco,sub
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../data/inflow_coxco.csv')
```

#cross check device count against EDM table
origin_fips = df['origin_fips'].unique().tolist()
origin_fips = pd.DataFrame(origin_fips,columns=['origin fips'])
origin_fips.to_csv('../data/inflow_unique-orig-fips.csv')

```python
#id 31cr counties and all others outside
df['origin_fips_reg'] = df['origin_fips'] 
df.loc[~df['origin_fips_reg'].isin(stco),'origin_fips_reg'] = 'O31CR' 
```

```python
df.head()
```

```python
dff = pd.pivot_table(df,values=['visits'],index=['date_start','origin_fips_reg'],columns=['destination_fips'],aggfunc=np.sum,fill_value=0,margins=True)
```
